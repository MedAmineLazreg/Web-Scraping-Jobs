```python
#!pip install selenium
from selenium import webdriver
import pandas as pd 
from bs4 import BeautifulSoup


driver = webdriver.Chrome("./chromedriver")


df = pd.DataFrame(columns=["Titre","Localisation","Entreprise","Salaire","Description"])

#Une boucle qui sert à parcourir n pages 
for i in range(0,5):
    
    #Effectuer une demande de l'url
    driver.get('https://www.indeed.fr/emplois?q=Data+Scientist&l=France&start='+str(i))
    
    #agrandir la fenetre
    driver.maximize_window()
    
    #Une attente implicite indique à WebDriver d'interroger
    #le DOM afin de trouver un élément (ou des éléments) qui ne sont pas immédiatement disponibles.  
    driver.implicitly_wait(20)
    

    for job in driver.find_elements_by_class_name('result'):
        
        #Spécifier un format souhaité de la page afin de lire ses différents composants un par un.
        soup = BeautifulSoup(job.get_attribute('innerHTML'),'html.parser')
        
        #Récuperer le titre, la localisation, le nom de l'entreprise, le salaire et les description de chaque offre,
        # en remplacant les retours (si trouvés) à la ligne avec des espace s
        try:
            titre = soup.find("a",class_="jobtitle").text.replace("\n","").strip()
            
        except:
            titre = 'Indisponible'

        try:
            localisation = soup.find(class_="location").text
        except:
            localisation = 'Indisponible'

        try:
            entreprise = soup.find(class_="company").text.replace("\n","").strip()
        except:
            entreprise = 'Indisponible'

        try:
            salaire = soup.find(class_="salary").text.replace("\n","").strip()
        except:
            salaire = 'Indisponible' 
            
        sum_div=job.find_elements_by_class_name("summary")[0]  
        
        #Utiliser une exception afin de fermer la fenêtre surgissante qui cause l'arret de notre programme
        try:
            sum_div.click()
        except:
            close_button = driver.find_elements_by_class_name('popover-x-button-close')[0]
            close_button.click()
            sum_div.click()

        desc = driver.find_element_by_id('vjs-desc').text

       
        df = df.append({'Titre':titre,'Localisation':localisation,"Entreprise":entreprise,"Salaire":salaire,
                        'Description':desc},ignore_index=True)

        print("L'avancement est parfait :",df.shape)


df.to_csv("Indeed.csv",index=False)
```
